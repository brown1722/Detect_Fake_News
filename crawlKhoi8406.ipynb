```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def content(link):
  response = requests.get(link)
  br_content = ''

  if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    cts = soup.find_all('br')
    for ct in cts:
            # Sử dụng .next_sibling để truy cập nội dung sau thẻ <br>
            next_sibling = ct.next_sibling

            # Kiểm tra xem next_sibling có phải là một chuỗi văn bản không trống
            if next_sibling and isinstance(next_sibling, str) and next_sibling.strip():
                br_content += next_sibling.strip() + " "
  else:
    print('error')
  return br_content
```

```python
df =[]
for page_number in range(1,213):
  url = f'https://www.khoi8406.com/?page={page_number}'

  response = requests.get(url)

  if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    posts = soup.find_all(class_='post-snippet')

    for post in posts:
      h2_tag = post.find('h2')

      href_value = post.find('a')['href'] if post.find('a') else None

      if h2_tag and href_value:
          df.append({'Title': h2_tag.text, 'URL': href_value,'Content': content(href_value), 'Label': '1'})
      data = pd.DataFrame(df)
  else:
    print("error")
```

```python
data.to_csv('../content/drive/MyDrive/PBL6/dataFakeNewsKhoi8406-4k.csv',index = False ,encoding='utf-8-sig')
```
